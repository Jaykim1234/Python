### Here I have made codes that allows us to get the lastest S&P 500 company makrket cap size rankings by sectors.
### I have picked top 3 large firms by sub-sectors. In order to get the lastest version of the S&P 500 company list, I have used Wikipedia as a source.

```python
import pandas as pd
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import time
import yfinance as yf
%matplotlib inline

# There are 2 tables on the Wikipedia page
# we want the first table

payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
first_table = payload[0]
second_table = payload[1]
df = first_table
```

```python
# This is the original S&P 500 table that we got from the Wikipedia

df.head()
```

```python
# As there is no market capital data, I have used Yahoofinance as a source to get the market captial data.

df['marketcap'] = df['Symbol'].apply(lambda x: yf.Ticker(x).info['marketCap'])
```

```python
# Since it takes a lot of time to add marketCap at the dataframe, I have exported this table as an exel form.

df.to_excel('df_with_marketcap.xlsx') 
df
```

```python
# Now I have sorted table by sub sectors and picked top 3 large firms .
df_rank3 = df.sort_values(['GICS Sector','marketcap','GICS Sub-Industry'], ascending=False).groupby('GICS Sub-Industry').head(3)[['Symbol','GICS Sector','GICS Sub-Industry','marketcap']]; 
df_rank3.to_excel('df_rank3.xlsx'); df_rank3
```

```python
df_rank10 = df.sort_values(['GICS Sector','marketcap','GICS Sub-Industry'], ascending=False).groupby('GICS Sub-Industry').head(10)[['Symbol','GICS Sector','GICS Sub-Industry','marketcap']]; 
df_rank10.to_excel('df_rank10.xlsx'); df_rank10
```

```python
df_rank10 = df.sort_values(['GICS Sub-Industry','marketcap'], ascending=False).groupby('GICS Sub-Industry').head(10)[['Symbol','GICS Sector','GICS Sub-Industry','marketcap']]; 
df_rank10.to_excel('df_rank10.xlsx'); df_rank10
```

```python
df_all_rank = df.sort_values(by=['GICS Sector', 'GICS Sub-Industry','marketcap'], ascending=[True, True, False])[['Symbol','GICS Sector','GICS Sub-Industry','marketcap']]; 
df_all_rank.to_excel('df_all_rank.xlsx'); df_all_rank
```
