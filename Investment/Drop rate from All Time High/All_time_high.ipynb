### Here I wrote a code that allows us to check how much index has decreased from All Time High value.
### Indexs that I have selected are SP500, DOW, NASDAQ, and RUSSEL 2000

```python
# Import the yfinance. If you get module not found error the run !pip install yfinance from your Jupyter notebook
!pip install yfinance
import yfinance as yf
import pandas as pd
```

```python
tickers_list = ['^DJI','^GSPC','^NDX','^RUT']

# Fetch the data
p_data = yf.download(tickers_list,'1900-1-1')['Adj Close']

# Print first 5 rows of the data
print(p_data.head())
print(p_data.tail())
```

```python
# Get the all time high value 
ath_dic = dict(p_data[['^DJI', '^GSPC', '^NDX', '^RUT']].max()); 

# Get the latest values
lastest_dic = dict(p_data.iloc[-1,:])

print(ath_dic) 
print(lastest_dic)
```

```python
df_all = pd.read_excel(r"C:\Users\Jinhyun\Documents\GitHub\Python\Investment\Drop rate from All Time High\ATH_v2.xlsx", sheet_name=None, header = 1)
df_ath = df_all['ATH']
display(df_ath)
```

```python
# Chnage name of columns 
df_ath.columns = ['Change', 'DOW', 'S&P 500', 'NASDAQ', 'RUSSELL2000', 'Situation']
display(df_ath)
```

```python
df_ath_chag = df_ath.iloc[1:201,:] 
display(df_ath_chag)
```

```python
# {'^DJI': 36799.6484375, '^GSPC': 4796.56005859375, '^NDX': 16573.33984375, '^RUT': 2442.739990234375}
df_ath_chag['DOW'] = (1+df_ath_chag['Change'])*ath_dic['^DJI']
df_ath_chag['S&P 500'] = (1+df_ath_chag['Change'])*ath_dic['^GSPC'] 
df_ath_chag['NASDAQ'] =(1+df_ath_chag['Change'])*ath_dic['^NDX'] 
df_ath_chag['RUSSELL2000'] = (1+df_ath_chag['Change'])*ath_dic['^RUT'] 

display(df_ath_chag)
```
